```python
CESdata = gpd.read_file('calenviroscreen40shpf2021shp.zip')
```

```python
CESdata.info()
```

```python
CESdata['Tract'] = CESdata['Tract'].astype(np.str)
CESdata['Tract'].info()
```

```python
CESdata['FIPS'] = '0'+CESdata['Tract']
CESdata.sample()
```

```python
CESdata['County'].value_counts()
```

```python
Kern_env = CESdata[CESdata.County == 'Kern']
Kern_env.info()
```

```python
Kern_env.sample(3)
```

```python
Kern_env = Kern_env[['TotPop19', 'ApproxLoc','Tract','FIPS','Hispanic','White','AfricanAm','NativeAm','OtherMult','Asthma','AsthmaP','Cardiovas','CardiovasP','LowBirtWt','LowBirWP','Shape_Leng','Shape_Area','geometry']]
Kern_env.head()
```

```python
Kern_env = Kern_env.to_crs("EPSG:4326") 
```

```python
fig,ax = plt.subplots(figsize=(12,10))
Kern_env.plot(ax=ax,facecolor='#ccc',edgecolor='white',lw=0.2)
```
